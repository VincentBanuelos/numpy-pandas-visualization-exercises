```python
import matplotlib.pyplot as plt
import numpy as np
```

1. Use matplotlib to plot the following equation:

y = x^2 - x + 2

You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, the origin.

```python
x = range(25)
y = [n ** 2 - n + 2 for n in x]
plt.plot(y)

plt.annotate('Origin (0,0)', xy=(0,0), xytext=(-0,100),
            arrowprops={'facecolor': 'blue'})

plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y = √x
 
y = x^3

y = 2^x

y = 1/(x+1)


```python
x = range(10)
y = [np.sqrt(n) for n in x]

plt.title('$y = \sqrt{x}$')
plt.xlabel('x')
plt.ylabel('y', rotation = 360)
plt.plot(x, y)
plt.show()
```

```python
y1 = [n ** 3 for n in x]
plt.plot(x, y1)
plt.title('$y = x^3$')
plt.xlabel('x')
plt.ylabel('y', rotation = 360)
plt.show()
```

```python
y2 = [2 ** n for n in x]
plt.plot(x, y2)
plt.title('$y = 2^n$')
plt.xlabel('x')
plt.ylabel('y', rotation = 360)
plt.show()
```

```python
y3 = [1/(n+1) for n in x]
plt.plot(x, y3)
plt.title('$y = 1/(x+1)$')
plt.xlabel('x')
plt.ylabel('y', rotation = 360)
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
# We can create multiple charts together using the .subplot function.
plt.subplot(2, 2, 1)
plt.title('$y = \sqrt{x}$')
plt.plot(x, y)

plt.subplot(2, 2, 2)
plt.title('$y = x^3$')
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x, y1)

plt.subplot(2, 2, 3)
plt.title('y = $2^x $')
plt.xlabel('x')
plt.plot(x, y2)

plt.subplot(2, 2, 4)
plt.title('$y = 1/(x+1)$')
plt.ylabel('y')
plt.plot(x, y3)


plt.tight_layout() # tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area. 
plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = range(5)

y  = [np.sqrt(n) for n in x]
y1 = [n ** 3 for n in x]
y2 = [2 ** n for n in x]
y3 = [1/(n+1) for n in x]
plt.figure(figsize=(14, 6)) # (width, height)

plt.plot(x, y, c='blue', label = '$y = \sqrt{x}$')
plt.plot(x, y1, c='red', label = '$y = x^3$')
plt.plot(x, y2, c ='orange', label = 'y = $2^x $')
plt.plot(x, y3, c ='green',label = '$y = 1/(x+1)$')

plt.legend(loc='upper left')
plt.title('4 Equations in 1')

plt.xlabel('x')
plt.ylabel('y', rotation = 360)

plt.show()
```
