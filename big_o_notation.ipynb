- Curves to graph
- y = 0n + 1 and label the curve "O(1)"

- y = n and label the curve "O(n)"
- y = n * log(n) and label it "O(log n)"
- y = n^2 and label it "O(n^2)"
- y = 2^n and label it "O(2^n)"
- y = n! and label it "O(n!)"
- y = n^n and label it "O(n^n)"

- y = 0n + 1 and label the curve "O(1)"

```python
plt.figure(figsize=(12,6))

x = range(1, 20)
y =  [(0 * n + 1) for n in x]
y1 = [math.log(n) for n in x]
y2 = [n for n in x]
y3 = [n * math.log(n) for n in x]
y4 = [math.pow(n,2) for n in x]
y5 = [math.pow(2,n) for n in x]
y6 = [math.factorial(n) for n in x]
y7 = [math.pow(n,n) for n in x]


plt.plot(x, y, label = "O(1)")
plt.plot(x, y1, label = "O(log(n))")
plt.plot(x, y2, label = "O(n)")
plt.plot(x, y3, label = "O(n log n))")
plt.plot(x, y4, label = "$O(n^2)$")
plt.plot(x, y5, label = "O($2^n$)")
plt.plot(x, y6, label = "O(n!)")
plt.plot(x, y7, label = "O($n^n$)")

plt.ylim(0,50)

plt.legend(loc="upper left")
plt.title("Big O Notation")
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.show()
```
