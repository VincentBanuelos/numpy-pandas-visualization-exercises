4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from env import host, user, password 

def get_db_url(database, user=user, host=host, password=password): 
    url = f'mysql+pymysql://{user}:{password}@{host}/{database}'
    return url
```

```python
url = get_db_url('chipotle')
orders = pd.read_sql('SELECT * FROM orders', url)
orders
```

```python
# Lets find the top 4 most ordered items first
top_four_items = orders.groupby('item_name').quantity.sum().sort_values(ascending=False).head(4)
top_four_items
```

```python
orders['item_price'] = orders['item_price'].str.strip('$').astype(float)
```

```python
item_revenue = pd.Series(orders.groupby(['item_name']).item_price.sum().sort_values(ascending=False))
item_revenue
```

```python
top_4_rev = pd.merge(top_four_items, item_revenue, left_on = 'item_name', right_on = 'item_name', how = 'inner')
top_4_rev
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x=top_4_rev.index, y='item_price', data=top_4_rev)
plt.title("Top 4 Item's Revenue")
plt.xlabel(' Item Name')
plt.ylabel('Revenue')
plt.show()
```

5. Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
sleep = data('sleepstudy')
data('sleepstudy', show_doc = True)
```

```python
# create data frame with data grouped by days that shows the average reaction time per day

avg_sleep = pd.DataFrame(sleep.groupby('Days').Reaction.agg('mean'))
```

```python
plt.figure(figsize=(12, 10))

# plot average for each day as line, set linewidth to 5 so it stands out, add label for legend, change color to black
plt.plot(avg_sleep.index, avg_sleep.Reaction, linewidth = 5, linestyle = '--', label = 'average', c = 'black')

# plot each patients reaction times, color by subject number, use palette that gives each patient a unique color
sns.lineplot(x ='Days', y='Reaction', hue = 'Subject', data = sleep, palette = 'Paired')

# move legend to outside of graph
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```
