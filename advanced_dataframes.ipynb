```python
mpg['average_mileage'] = (mpg['cty'] + mpg['hwy'])/2
mpg
```

14. Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.

```python
mpg['stick_or_nah'] = mpg['trans'].apply(lambda x : x[0:4] if x[0:4] == 'auto' else x[0:6])

#seeing that in the trans column it does indicate if they are an auto or not, however it is ont in a format that will be useful to create the is automatic column,
#the lambda function above goesthrough the first 5 characters and see if they equal auto, if so, stick or nah will display auto, otherwise, it will display manual.
```

```python
mpg['is_automatic'] = mpg['stick_or_nah'] == 'auto'
```

```python
mpg
```

15. Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?
